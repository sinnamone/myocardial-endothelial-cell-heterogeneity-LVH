```python
import scanpy as sc
import seaborn as sns
```

```python
adata = sc.read("Part4_cell_annotation.h5ad")
```

```python
adata
```

```python
sc.tl.dendrogram(adata, "cell_annoation",cor_method="spearman")
```

```python
sc.set_figure_params(dpi=150)
sc.settings.set_figure_params(dpi=100, facecolor='white')
```

```python
ax = sc.pl.dendrogram(adata, "cell_annoation")
```

```python
sc.set_figure_params(dpi=100)
sns.set_theme(style='white')
ax = sc.pl.correlation_matrix(adata, "cell_annoation", figsize=(6, 6),save="correlation_matrix_spearman.pdf")
```
